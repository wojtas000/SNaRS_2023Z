## 1st version

```python
def draw_lattice(n, m):
    for i in range(n):
        for j in range(m):
            
            if j != m-1:
                print("O", "--", end=" ")
            else:
                print("O")
        
        if i != n-1:
            
            for j in range(m):
                if j != m-1:
                    print("|", "  ", end=" ")
                else:
                    print("|")
```

```python
draw_lattice(5, 8)
```

# Task 3

```python
def complete_graph(num_points):
    radius = 1.0

    angles = np.linspace(0, 2 * np.pi, num_points, endpoint=False)

    x_coordinates = radius * np.cos(angles)
    y_coordinates = radius * np.sin(angles)
    edges = [(x_coordinates[i], y_coordinates[i], x_coordinates[j], y_coordinates[j]) for i in range(num_points) for j in range(num_points) if i != j]

    fig = plt.figure(figsize=(6, 6))

    plt.scatter(x_coordinates, y_coordinates, color='skyblue', s=100)

    for edge in edges:
        plt.plot([edge[0], edge[2]], [edge[1], edge[3]], color='black')

    plt.xlim(-1.5, 1.5)
    plt.ylim(-1.5, 1.5)
    plt.title(f'Complete graph with {num_points} nodes')
    plt.show()
```

```python
complete_graph(7)
```

# Task 4

```python
G = nx.Graph()
G.add_edge('A', 'B', weight=4)
G.add_edge('B', 'D', weight=2)
G.add_edge('A', 'C', weight=3)
G.add_edge('C', 'D', weight=4)
G.add_edge('C', 'E', weight=5)
G.add_edge('D', 'E', weight=1)
G.add_edge('D', 'F', weight=6)
G.add_edge('B', 'C', weight=2)
```

```python
node_colors = ['blue', 'green', 'red', 'yellow', 'orange', 'purple']

pos = nx.spring_layout(G)
nx.draw(G, pos, with_labels=True, node_color=node_colors, cmap=plt.cm.Reds, node_size=1000)
labels = nx.get_edge_attributes(G, 'weight')
nx.draw_networkx_edge_labels(G, pos, edge_labels=labels)
plt.show()
```
