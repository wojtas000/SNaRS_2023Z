```python
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np

# Barabasi-Albert (BA) network
n = 500
m = 3  
ba_graph = nx.barabasi_albert_graph(n, m)

# Erdos-Renyi (ER) random graph
p = 0.012
er_graph = nx.erdos_renyi_graph(n, p)

# Calculate degree distributions
ba_degrees = [degree for node, degree in ba_graph.degree()]
er_degrees = [degree for node, degree in er_graph.degree()]


# Plot degree histograms
plt.figure(figsize=(12, 6))

plt.subplot(121)
plt.hist(ba_degrees, bins=12, color='b', alpha=0.6, label='BA Network')
plt.xlabel('Degree')
plt.ylabel('Count')
plt.title('Degree Histogram (BA Network)')
plt.legend()

plt.subplot(122)
plt.hist(er_degrees, bins=12, color='r', alpha=0.6, label='ER Random Graph')
plt.xlabel('Degree')
plt.ylabel('Count')
plt.title('Degree Histogram (ER Random Graph)')
plt.legend()

plt.tight_layout()
plt.show()

ba_hubs = [node for node, degree in ba_graph.degree() if degree >= np.percentile(ba_degrees, 90)]
er_hubs = [node for node, degree in er_graph.degree() if degree >= np.percentile(er_degrees, 90)]

print(f'BA Network has {len(ba_hubs)} hubs (top 10%).')
print(f'ER Random Graph has {len(er_hubs)} hubs (top 10%).')

ba_avg_degree = np.mean(ba_degrees)
er_avg_degree = np.mean(er_degrees)

print(f'Average degree in BA Network: {ba_avg_degree:.2f}')
print(f'Average degree in ER Random Graph: {er_avg_degree:.2f}')
```

```python
plt.figure(figsize=(12, 6))
pos = nx.spring_layout(ba_graph)
# color by degree
cmap = plt.cm.plasma(ba_degrees / np.max(ba_degrees))

nx.draw(ba_graph, pos, node_size=10*np.array(ba_degrees), node_color=cmap, alpha=0.8, with_labels=False, width=0.1)
plt.show()
```

```python
# draw the ba_graph with networkX
plt.figure(figsize=(12, 6))
pos = nx.spring_layout(er_graph, k=0.2)
# color by degree
cmap = plt.cm.plasma(er_degrees / np.max(er_degrees))

nx.draw(ba_graph, pos, node_size=10*np.array(er_degrees), node_color=cmap, alpha=0.8, with_labels=False, width=0.1)
plt.show()
```
